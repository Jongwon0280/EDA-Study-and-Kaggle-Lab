```python
from selenium import webdriver

# Chrome WebDriver 경로를 설정합니다. 다운로드 링크: https://sites.google.com/a/chromium.org/chromedriver/downloads
webdriver_path = 'chromedriver'

# Chrome WebDriver 실행
browser = webdriver.Chrome(webdriver_path)

url = "https://taas.koroad.or.kr/gis/mcm/mcl/initMap.do?menuId=GIS_GMP_STS_RSN"
browser.implicitly_wait(3)
browser.get(url)  # 페이지 접속


#browser.quit()  # 브라우저 종료
```

```python
from selenium.webdriver.support.ui import Select
```

> ### 4.1 웹페이지에서 원하는 조건을 전달

<br>

1. 콤보박스를 통해 원하는 지역 및 동을 선택

2. 체크박스를 통해 원하는 사고유형 (all, 사망, 중상, 경상)을 선택

3. 버튼클릭으로 조건에 맞는 목록을 불러온다.

```python
#시작 및 끝 년도를 입력합니다.

startYear = ""

EndYear=""


# 콤보박스 선택
combo_box = browser.find_element(By.ID,"ptsRafYearStart")  # 콤보박스 요소를 식별합니다.

yearStart = Select(combo_box)

yearStart.select_by_visible_text(startYear)

combo_box = browser.find_element(By.ID,"ptsRafYearEnd")

yearEnd = Select(combo_box)

yearEnd.select_by_visible_text(EndYear)

#시 도
# "28" => 인천
combo_box = browser.find_element(By.ID,"ptsRafSido")

sido = Select(combo_box)

sido.select_by_value("28")


#구 동
# :: 전체 ::  -> "01"

combo_box = browser.find_element(By.ID,"ptsRafSigungu")

dong = Select(combo_box)

dong.select_by_value("01")
```

```python
# 체크박스
# 사망사고는 default로 체크되어있음. 그래서 2부터 4까지 (모든 사고유형을 다 포함시킨다.)
for i in range(2,4+1):
    

    check_box = browser.find_element(By.XPATH ,f'//*[@id="ptsRafCh1AccidentContent"]/li[{i}]/input')

    check_box.click()

```

```python
# 조회버튼

search_button = browser.find_element(By.XPATH ,'//*[@id="regionAccidentFind"]/div[2]/p/a')

search_button.click()

# 목록 불러오는데 시간이 걸리기에 wait 걸기
browser.implicitly_wait(3)


# 목록보기 버튼

search_button = browser.find_element(By.XPATH ,'//*[@id="regionAccidentFind"]/div[3]/div[2]/p/a')

search_button.click()
```

```python
# 새로운 창이뜨면, 그창으로 핸들을 옮긴다.

new_window = browser.window_handles[-1]  # 가장 최근에 열린 창의 핸들을 가져온다.
browser.switch_to.window(new_window)

```

```python
col_name=[i for i in range(0,17+1)] # 테스트로 컬럼명은 그냥 정수로 지정했음 18개 컬럼
```
